# Tratamento de dados

```python
import csv
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize 
from wordcloud import WordCloud
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
import unicodedata

#colocando todo o tweet em maiusculo
df['content'] = df['content'].str.upper() 

#função de remover acentos
def remove_combining_fluent(string: str) -> str:
    normalized = unicodedata.normalize('NFD', string)
    return ''.join([l for l in normalized if not unicodedata.combining(l)]
    )

df['content'] = df['content'].apply(remove_combining_fluent)
df
```

```python
df
```

```python
#criando os caracteres que serão removidos
stop_words = nltk.corpus.stopwords.words('portuguese')
stop_words = stop_words + list([ 'ESSE','QUEM','PELO','MESMO','ESSA','SOBRE','ISSO','ESTA',',','.','HTTPS','HTTP', ':','!',
'#', '@', '?', '/', 'O', 'VAI', '|', '–','PRA', 'A', 'E', '-','"',"'", 'PARA', 'MAIS', 'COM', 'SE', 'UM', 'UMA', 'DE', 'DO',
 'DA', 'OS', 'AS', 'EM', 'NO', 'NA', 'NOS', 'NAS', 'POR', 'MAS', 'COMO', 'MAS', 'AO', 'PELA', 'PELOS', 'PELAS', 'E', 'SAO', 
 'TAMBEM', 'PODEMOS', 'PODEIS', 'PODEM', 'PODES', 'PODE', 'PODIA', 'PODIAM', 'PODIAIS', 'PODIAMOS', 'PO'])
 
def remove_stopwords(x):
  word_tokens = nltk.word_tokenize(x)
  # print(word_tokens)
  filtered_sentence = ' '.join([w for w in word_tokens if not (w in stop_words or len(w)<4)]) 
  return filtered_sentence

#criando a coluna filtered_text com os tweets sem as stopwords
df['filtered_text'] = df['content'].apply(lambda x: remove_stopwords(x))

#apagando a coluna content
df.drop(["content"], axis=1, inplace=True)
```

```python
df.to_csv( "tweetsDe2012a2023Stopwords.csv", index=False, encoding='utf-8-sig')
```

```python
df = pd.read_csv('tweetsDe2012a2023Stopwords.csv', )
```

```python
#dividindo os tweets em palavras
df['tokens'] = df.apply(lambda row: nltk.word_tokenize(row['filtered_text']), axis=1)
```

```python
#dividindo os tweets em bigramas
from nltk import bigrams
df['bigramas'] = df.apply(lambda row: list(bigrams(row['tokens'])), axis=1)
```

```python
#dividindo os tweets em trigramas
from nltk import trigrams
df['trigramas'] = df.apply(lambda row: list(trigrams(row['tokens'])), axis=1)
```
